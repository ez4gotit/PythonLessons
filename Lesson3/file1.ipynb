### Типы и Классы

```python
#Типы vs Класса

#Тип это некотрая структура представления данных, не  имеющая внутренней логики взаимодействия

a : str = "b"

c :str = "d"

#f(x) = y = x+1

def f(x):
    y = x+1
    return y
    

def concatinate_strings(a, b):
    return a + b

e = concatinate_strings(a,c)
print(e)


#Класс это некоторый набор параметров, свойств, и методов взаимодействия объектов с другими объектами и типами данных для решения задач


class Person:

    def __init__(self, name : str, age : int):
        self.name = name
        self.age = age

    def printName(self):
        print(self.name)


    def printFullInfo(self):
        print(f'Name: {self.name} \n Age: {self.age}')

    def ageDifference(self, otherPerson) -> float:
        return float(otherPerson.age-self.age)


person1 = Person(name =  "AAAA", age = 8)
person2 = Person(name = "name",  age = 2)


print(person1.ageDifference(person2))

```
