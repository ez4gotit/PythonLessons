```python

#Немного классов и ООП:
class Person():
    def __init__(self, name, age, gender, videogame):
        self.name = name
        self.age = age
        self.gender = gender
        self.videogame = videogame
        #В конструкторе  может быть логика
        if videogame == "Дота2":
            print("No friends :(")


    #Метод
    def print(self):
        print(self.name)
        print(self.age)
        print(self.gender)
        print(self.videogame)

name = input("Введите Имя")
age = input("Введите возраст")
gender = input("Vvedite gender")
game = input("Напишите Дота2")

a = Person(name=name,age=age, gender=gender,videogame=game)
b = a

#bimbimbambam
print(b.videogame)

#Norm variant

a.print()
```
